```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('merged_movie_data.csv')

# Split genres and calculate average IMDB score
data['genres'] = data['genres'].str.split('|')
genre_scores = data.explode('genres').groupby('genres')['imdb_score'].mean().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10, 6))
genre_scores.plot(kind='bar', color='skyblue')
plt.title('Average IMDB Score by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average IMDB Score', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


#1. **Genre Popularity**: By analyzing the average IMDB scores, we can identify which genres are most popular or highly rated among audiences.
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('merged_movie_data.csv')
print(df.head())

# Fill missing values
df['budget_x'].fillna(df['budget_x'].mean(), inplace=True)
df['runtime'].fillna(df['runtime'].mean(), inplace=True)
df['imdb_score'].fillna(df['imdb_score'].mean(), inplace=True)

# Convert categorical data to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['genres', 'country', 'certification_US'], drop_first=True)

# Define input features (X) and target variable (y)
X = df[['budget_x', 'runtime', 'imdb_score', 'vote_average', 'vote_count'] +
       [col for col in df.columns if 'genres_' in col or 'country_' in col or 'certification_US_' in col]]
y = df['success'].astype(int)  # Ensure the target is in numerical format

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Building: Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Predict success for new movies (example)
new_movie = pd.DataFrame({
    'budget_x': [200000000],
    'runtime': [120],
    'imdb_score': [8.0],
    'vote_average': [7.5],
    'vote_count': [2000],
    # Set genre and other categorical dummy variables
    'genres_Action': [1], 'genres_Adventure': [1], 'genres_Sci-Fi': [0],  # Adjust as per your genres
    'country_United States of America': [1],
    'certification_US_PG-13': [1]
}, index=[0])

new_movie = new_movie.reindex(columns=X.columns, fill_value=0)  # Ensure correct column order
success_prediction = model.predict(new_movie)
print("\nPredicted Success:", "Successful" if success_prediction[0] == 1 else "Not Successful")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('merged_movie_data.csv')


df.fillna({'director_name': 'Unknown', 'actor_1_name': 'Unknown', 'actor_2_name': 'Unknown', 'actor_3_name': 'Unknown'}, inplace=True)
df.fillna(0, inplace=True)  # Replace numeric NaNs with 0 for simplicity

# Step 3: Calculate Director Success Rate
director_success = df.groupby('director_name')['success'].mean().to_dict()
df['director_success_rate'] = df['director_name'].map(director_success).fillna(0)

# Step 4: Calculate Actor Success Rates
for actor_col in ['actor_1_name', 'actor_2_name', 'actor_3_name']:
    actor_success = df.groupby(actor_col)['success'].mean().to_dict()
    df[f'{actor_col}_success_rate'] = df[actor_col].map(actor_success).fillna(0)

# Step 5: Calculate Genre Success Rate
df['genres'] = df['genres'].fillna('Unknown')  # Fill missing genres with 'Unknown'
genre_success = df.explode('genres').groupby('genres')['success'].mean().to_dict()
df['genre_success_rate'] = df['genres'].apply(
    lambda g: np.mean([genre_success.get(genre.strip(), 0) for genre in g.split('|')])
).fillna(0)

# Step 6: Feature Selection
features = [
    'imdb_score', 'budget_x', 'vote_average', 'runtime',
    'director_success_rate', 'actor_1_name_success_rate', 'actor_2_name_success_rate',
    'actor_3_name_success_rate', 'genre_success_rate'
]
X = df[features]  # Feature matrix
y = df['success'].astype(int)  # Target variable

# Step 7: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 8: Model Training
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 9: Model Evaluation
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{classification_rep}")

# Step 10: Prediction Function for a New Movie
def predict_new_movie(data):
    """
    Predict the success of a new movie based on the input data.

    Parameters:
        data (dict): Dictionary containing the attributes of the movie.

    Returns:
        str: Prediction ("Successful" or "Unsuccessful").
    """
    # Fill missing fields with default values
    data = {
        'imdb_score': data.get('imdb_score', df['imdb_score'].mean()),
        'budget_x': data.get('budget_x', df['budget_x'].mean()),
        'vote_average': data.get('vote_average', df['vote_average'].mean()),
        'runtime': data.get('runtime', df['runtime'].mean()),
        'director_name': data.get('director_name', 'Unknown'),
        'actor_1_name': data.get('actor_1_name', 'Unknown'),
        'actor_2_name': data.get('actor_2_name', 'Unknown'),
        'actor_3_name': data.get('actor_3_name', 'Unknown'),
        'genres': data.get('genres', 'Unknown'),
    }

    input_data = pd.DataFrame([data])

    # Calculate success rates
    input_data['director_success_rate'] = input_data['director_name'].map(director_success).fillna(0)
    for actor_col in ['actor_1_name', 'actor_2_name', 'actor_3_name']:
        input_data[f'{actor_col}_success_rate'] = input_data[actor_col].map(actor_success).fillna(0)
    input_data['genre_success_rate'] = input_data['genres'].apply(
        lambda g: np.mean([genre_success.get(genre.strip(), 0) for genre in g.split('|')])
    ).fillna(0)

    # Select relevant features
    input_data = input_data[features]

    # Predict
    prediction = model.predict(input_data)
    return "Successful" if prediction[0] == 1 else "Unsuccessful"

# Example Prediction
new_movie = {
    'budget_x': 100000000,
    'runtime': 120,
    'director_name': 'Christopher Nolan',
    'actor_1_name': 'Christian Bale',
    'actor_2_name': 'Michael Caine',
    'actor_3_name': 'Unknown',
    'genres': 'Action|Adventure'
}

prediction = predict_new_movie(new_movie)
print("Prediction for the new movie:", prediction)
```
